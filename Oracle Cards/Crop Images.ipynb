## This is for resizing/processing images to be 1024 x 1024! ##

This is necessary, since the Tarot Cards are portrait-oriented and I need them to be in a square format for the StyleGAN to be trained in the most accurate manner possible.

```python
import cv2
import numpy as np
import os

## This rescales the image to fit on 1024 x 1024 and adds white space as needed

count = 0

for entry in os.listdir('./card_images/second_batch_images_for_ML'):
    if entry.endswith('.jpg'):
        
        try:
            img = cv2.imread('./card_images/second_batch_images_for_ML/' + entry, cv2.IMREAD_UNCHANGED)
            print('original dimension: ', img.shape)
            if img.shape[0] > img.shape[1]: #double check for portrait mode, cuz some aren't and those are noise
                scale_percent = 1024 / img.shape[0]
                width = int(img.shape[1] * scale_percent)
                height = int(img.shape[0] * scale_percent)
                dim = (width, height)
                resized = cv2.resize(img, dim, interpolation = cv2.INTER_AREA)
                image = cv2.copyMakeBorder(resized, 0, 0, int((1024 - width)/2), int((1024 - width)/2), cv2.BORDER_CONSTANT,value=[255,255,255])
                res = cv2.imwrite('./card_images/processed/'+ str(entry), image)
                
    
                
            else:
                print(entry + ' was not processed')
            
#             scale_percent = 
            count += 1
            print (count)
        
        except Exception as e: ## This gives you error message
            print ('could not process ' + entry)
            print (e)
            
```
